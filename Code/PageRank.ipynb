# <center> Can PageRank predict ranking of influential users?

```python
import numpy as np
```

Load data: n_retweets[uid,rtid] = nb times uid has retweetd rtid.

```python
n_retweets = dict()
for line in open("../Newman/trace.txt"):
    line = line.split()
    uid, rtid = int(line[2]), int(line[-1])
    if rtid != -1:
        if uid in n_retweets:
            if rtid in n_retweets[uid]:
                n_retweets[uid][rtid] += 1
            else:
                n_retweets[uid][rtid] = 1
        else:
            n_retweets[uid] = dict()
            n_retweets[uid][rtid] = 1      
```

Random walk.

```python
n_iter = 150000 # choose number of iterations

retweeters = np.fromiter(n_retweets.keys(), dtype=int)
uid = np.random.choice(retweeters)
score = dict()

for i in range(n_iter):
    
    # update current user's score
    if uid not in score:
        score[uid] = 1
    else:
        score[uid] += 1
    
    # if current user has not retweeted
    if uid not in retweeters:
        uid = np.random.choice(retweeters)
    
    # otherwise move to a neighbor according to nb retweets
    else:
        neighbors = np.fromiter(n_retweets[uid].keys(), dtype=int)
        weights = np.fromiter(n_retweets[uid].values(), dtype=int)
        weights = weights / weights.sum()
        uid = np.random.choice(neighbors, p=weights)
```

Print results.

```python
for uid, score in sorted(score.items(), key=lambda x:x[1], reverse=True):
    print(uid, score)
```
