### R√©cup√©rer infos utiles
- `createdAt_to_tuple`: fonction utilitaire pour transformer date des tweets en timestamp. /!\ pbs de timezone
- `tweet_to_tuple`: transforme un tweet extended (dict ou str) en tuple (twid, uid, ts, rtid, rtuid).

```python
def createdAt_to_timestamp(s):
    
    # recup infos
    month = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06', 
             'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
    d = s[8:10]
    m = month[s[4:7]]
    Y = s[-4:]
    H = s[11:13]
    M = s[14:16]
    S = s[17:19]
    z = s[-10:-5]
    
    # to timestamp
    date = d + '/' + m + '/' + Y + '-' + H + ':' + M + ':' + S + '+' + z
    ts = time.mktime(datetime.datetime.strptime(date, "%d/%m/%Y-%H:%M:%S+%z").timetuple())
    
    # return
    return int(ts)
```

```python
def tweet_to_tuple(tweet):

    # si string on tansforme en dict
    if type(tweet)==str:
        tweet = json.loads(tweet)
        
    # timestamp
    if 'timestamp_ms' in tweet.keys():
        ts = int(round(int(tweet['timestamp_ms'])/1000, 0))
    else:
        ts = createdAt_to_timestamp(tweet['created_at'])
    
    # uid, twid
    twid = tweet['id']
    uid = tweet['user']['id']

    # retweet
    if 'retweeted_status' in tweet.keys():
        rtid = tweet['retweeted_status']['id']
        rtuid = tweet['retweeted_status']['user']['id']
    else:
        rtid = -1
        rtuid = -1

    # return
    return (twid, uid, ts, rtid, rtuid)
```

```python
tweet = twitter.show_status(id=850866638400479234, tweet_mode='extended')
tweet_to_tuple(tweet)
```
